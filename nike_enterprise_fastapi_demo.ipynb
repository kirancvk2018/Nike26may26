## Refactoring, Test Generation, Documentation with Agents

```
You are an enterprise AI coding agent working in Cursor.

Context:
- FastAPI project: Nike Inventory Service
- Files: app/services/inventory_service.py, app/repository/inventory_repo.py

Tasks:
1. Refactor inventory_service.py:
   - Add input validation
   - Improve naming and structure
   - Keep business logic intact

2. Generate pytest tests in app/tests/test_inventory.py:
   - Valid item
   - Invalid quantity
   - Missing fields

3. Add docstrings and comments

Constraints:
- Small diffs only
- Follow PEP8
- Use only existing files

Output:
- Refactored code
- Updated test file
```

### Explanation

**Impact:**
Improves maintainability and introduces test coverage in a real service layer.

**Pros:**
- Realistic enterprise refactoring
- Demonstrates AI-assisted TDD
- Works directly on service + repo layers

**Cons:**
- Limited by in-memory DB
- No async or DB validation yet

**Expected Output:**
- Cleaner service logic
- Expanded pytest cases
- Proper documentation added

## Single-Agent vs Multi-Agent Workflows

```
You are designing AI workflows.

Context:
- FastAPI project
- Flow: Add inventory via API → store in repo

Tasks:
1. Single-agent:
   - Update API + service to validate quantity > 0

2. Multi-agent:
   - Planner: define validation
   - Executor: update service
   - Reviewer: update tests

Constraints:
- Modify only existing files

Output:
- Code changes
- Workflow comparison
```

### Explanation

**Impact:**
Shows real difference between simple vs orchestrated AI workflows.

**Pros:**
- Clear enterprise mapping
- Demonstrates collaboration pattern

**Cons:**
- Requires explanation during demo
- Not fully automated orchestration

**Expected Output:**
- Updated validation logic
- Improved tests
- Clear workflow breakdown

## Eval-Driven Cursor Usage — Prompts, Specs, Agent Validation

```
You are implementing evaluation-driven development.

Context:
- File: app/services/inventory_service.py

Tasks:
1. Add validation:
   - quantity must be > 0
   - sku must not be empty

2. Define spec:
   - valid input
   - invalid quantity
   - missing sku

3. Validate using pytest

Output:
- Updated code
- Spec
- Validation results
```
